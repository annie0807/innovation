```python
import pandas as pd
import math
#df = pd.read_excel('樞紐分析表_shannon.xlsx','CN_ACname樞紐分析表')
df = pd.read_excel('樞紐分析表_shannon_622.xlsx','WIPO_AC樞紐分析表')
```

```python
col_name = []
for i in df.iloc[0]:
    col_name.append(i)
df.set_axis(col_name, axis=1, inplace = True)
df.drop([0],inplace = True)
```

```python
row_name = []
for i in range(len(df.iloc[:,0])):
    row_name.append(df.iloc[i,0])
df.set_axis(row_name, axis=0, inplace = True)
df.drop(columns = ['列標籤'], inplace = True)
```

```python
df
```

```python
import numpy as np
def shannon(pi):
    index = []
    for x in pi:
        p = -1 * (x * math.log(x))
        index.append(p)
    Sindex = np.nansum(index)
    return Sindex
```

```python
shannons = []
for j in range(len(df.iloc[:,0])):
    pi = []
    for i in range(len(df.iloc[j,:])):
        if i != len(df.iloc[j, :])-1:
            pi.append(df.iloc[j, i] / df.iloc[j, -1])
    shannons.append(shannon(pi))
```

```python
print(len(shannons))
```
